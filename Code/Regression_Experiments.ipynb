```python
ff = pd.read_csv(data_dir + 'Fama_french.csv')
ff.dropna(inplace=True)
ff.yyyymm = ff.yyyymm.astype(int)
ff.rename(columns={'Mkt-RF':'mkt_rf'},inplace=True)
ff['date'] = pd.to_datetime(ff['yyyymm'],format='%Y%m')+pd.DateOffset(months=-1)
returns = pd.read_parquet(data_dir + 'processed_returns.parquet')
```

```python
imputed = os.path.join(data_dir, "imputed/*.parquet")
all_files = glob.glob(imputed)
dfs = {}
for f in all_files:
    filename = os.path.splitext(os.path.basename(f))[0]
    print(filename)
    dfs[filename] = pd.read_parquet(f)
```

```python
def resample(df):
    df.loc[df.s==2,'s'] = 3
    df.loc[df.s==1,'s'] = 2
    df.loc[(df.year>2010)&(df.year<2013),'s'] = 1
    return df
```

```python
rets = ['1m.lret','6m.lret','12m.lret']
for f,df in dfs.items():
    df = resample(df)
    df = pd.merge(df,returns,on=['yyyymm','permno'],how='left')
    dfs[f] = df
```

```python
combinations1 = []
for f,df in dfs.items():
    for r in rets:
        t = {'name':f+'__'+r[:2],'data':df,'vars': df.columns[df.columns.str.contains('z')],'yvar':r,'train_freq':'1m','turnover':'month'}
        combinations1.append(t)
```

```python
combinations2 = []
for f,df in dfs.items():
    for r in rets:
        t = {'name':f+'__'+r[:2],'data':df,'vars': df.columns[df.columns.str.contains('z')],'yvar':r,'train_freq':'12m','turnover':'year'}
        combinations2.append(t)
```

```python
combinations4 = []
for f,df in dfs.items():
    for r in rets:
        t = {'name':f+'__'+r[:2],'data':df,'vars': df.columns[df.columns.str.contains('z')],'yvar':r,'train_freq':'1m','turnover':'year'}
        combinations3.append(t)
```
